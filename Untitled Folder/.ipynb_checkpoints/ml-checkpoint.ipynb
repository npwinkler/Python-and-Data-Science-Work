```python
def J(X,theta,y):
    
    #Convert to np.matrix format if not already
    #np.matrix is essentially a 2-d np.array, but instead of vectorization it uses
    #matrix operations (for multiplication)
    X = np.asmatrix(X)
    theta = np.asmatrix(theta)
    y = np.asmatrix(y)
    m = X.shape[0]
    # You can use np.matrix() instead of np.asmatrix().  Find out why np.asmatrix()
    # is better.
    
    #Check that matrix shapes correspond 
    if theta.shape[0] != X.shape[1] or X.shape[0] != y.shape[0] or y.shape[1] != 1:
        print("Incompatible shapes:", theta.shape, X.shape, y.shape)
        raise Exception
        
    h = X * theta
    
    # J below doesn't use elementwise operations (vectorization), rather matrix operations
    # J = ((h - y).T * (h-y))/(2.0 * m)
    
    # The following doesn't work as it assumes vectorization
    # J = ((h-y)**2).mean()/2.0
    
    # But the following coverts back to np.array format and so works
    J1 = (np.asarray(h-y)**2).mean(axis=0)/2.0
    # It is better to work with "average cost" rather than "total cost" because 
    # then the learning rate doesn't have to be tuned for different number of 
    # examples
    
    # Check that both matrix based and vectorization based mechanisms give the same
    # answer
    # if abs(J-J1) >= 0.00001*J:
    #    print(J, J1)
    #    raise Exception
    
    return J1
```

```python
# Create the X matrix we'll be using, including the first column of 1s

X = np.append(np.ones((len(df),1)), np.asmatrix(df[['area', 'bedrooms']]), axis=1)
print(X.shape)
print(X[:5,:])
```

```python
# Create the y vector

y = np.asmatrix(df.price).T
print(y.shape)
```

```python
# Plot the contours of J for varying theta0 and theta1

theta0s = np.linspace(50000,130000,200) # not normalized
theta1s = np.linspace(120,160,200) # not normalized
t0_varying, t1_varying = np.meshgrid(theta0s, theta1s)
pos = np.empty(t0_varying.shape + (3,))
pos[:,:,0] = t0_varying
pos[:,:,1] = t1_varying
pos[:,:,2] = -8700
pos_2d = pos.reshape((-1,3))

cost = J(X, pos_2d.T, y)
print X.shape, pos_2d.T.shape, y.shape
# Here X is a 47 x 3 matrix, y is a 47 x 1 matrix, as expected.
# But pos_2d.T is a 3 x 40,000 matrix, representing all the 200x200
# theta0, theta1 pairs we want to plot J() for.  The fascinating part
# (which wasn't explicitly mentioned in class in Week 1) is that the vectorized
# implementation of J can handle this as well.

cost = cost.reshape(t0_varying.shape)

plt.figure()
plt.contour(t0_varying, t1_varying, cost, np.arange(2,2.2,0.005)*1e9)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(t0_varying, t1_varying, cost)
```

```python
plt.figure()
theta0s = np.linspace(-20000,160000, 3000) # not normalized
costs = []
for t0 in theta0s:
    costs.append(J(X, np.matrix([t0, 139, -8400]).T, y))
plt.plot(theta0s, costs)
```

```python
plt.figure()
theta1s = np.linspace(100, 200, 1000) # not normalized
costs = []
for t1 in theta1s:
    costs.append(J(X, np.matrix([89000, t1, -8400]).T, y))
plt.plot(theta1s, costs)
```
